```python
import numpy as np

celsius_q = np.array([-40, -10,  0,  8, 15, 22,  38],  dtype=float)
fahrenheit_a = np.array([-40,  14, 32, 46, 59, 72, 100],  dtype=float)

for i, c in enumerate(celsius_q):
  print("{} degrees Celsius = {} degrees Fahrenheit".format(
      c, fahrenheit_a[i]))
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

celsius_q = torch.tensor(celsius_q.reshape(-1, 1), dtype=torch.float32)
fahrenheit_a = torch.tensor(fahrenheit_a.reshape(-1, 1), dtype=torch.float32)
```

```python
model = nn.Sequential(nn.Linear(1, 1))
loss_func = nn.MSELoss()
optimizer = optim.Adam(model.parameters(), lr=0.1)
```

```python
n_epochs = 5000

for epoch in range(n_epochs):
    # Forward pass
    y_pred = model(celsius_q)

    # Compute Loss
    loss_val = loss_func(y_pred, fahrenheit_a)

    # Zero gradients, perform a backward pass, and update the weights.
    optimizer.zero_grad()
    loss_val.backward()
    optimizer.step()

    print(f'Epoch {epoch+1}, loss {loss_val.item()}')
```
